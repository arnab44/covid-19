```python
dataset1 = pd.read_csv('../covid-data-india-3.csv')
```

```python
dataset1.tail()
```

```python
dates = dataset1.loc[:, 'date'].values;
total_infectives = dataset1.loc[:, 'totalconfirmed'].values;
total_death = dataset1.loc[:, 'totaldeceased'].values;
total_recovered = dataset1.loc[:, 'totalrecovered'].values;
total_removed = total_death + total_recovered;
net_infectives = total_infectives - total_removed;
```

```python
plt.rcParams["figure.figsize"] = (100,40)
plt.plot(dates, net_infectives, label = 'net infectives')
plt.plot(dates, total_infectives, color='g', label = 'total infectives')
plt.plot(dates,total_death, color = 'r', label = 'total death')
plt.plot(dates,total_recovered, color ='b', label= ' total recovered')
plt.title('Time analysis')
plt.xlabel('Date')
plt.ylabel('Rate')
plt.legend()
plt.show()
```

```python
dr_dt = np.gradient(total_removed)
dr_dt
```

```python
plt.plot(dates,dr_dt)
plt.plot(dates, total_recovered, color='g')
```

```python
a = dr_dt/net_infectives
```

```python
plt.plot(dates,a)
```

```python
s0 = total_population * np.ones(98) 
s = s0- net_infectives - total_removed
plt.plot(dates, s)
plt.title('sussiptable ove time')
```
